```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cpi
```

```python
cpi.update()
```

```python
# access the data

df = pd.read_csv('C:/Users/reezal93/Documents/Data Analyst/python_movies_portfolio/python_movies.csv')
df.head()
```

```python
# make a list of missing data percentages for each feature

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# get a summary of the data frame

df.info()
```

```python
# remove missing values

df_clean = df.dropna()
df_clean.head()
```

```python
# make a list of missing data percentages for each feature

for col in df_clean.columns:
    pct_missing = np.mean(df_clean[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# get a summary of the data frame

df_clean.info()
```

```python
# change data type for one or more columns

df_clean['budget'] = df_clean['budget'].astype('int64')
df_clean['gross'] = df_clean['gross'].astype('int64')
df_clean['votes'] = df_clean['votes'].astype('int64')
df_clean['runtime'] = df_clean['runtime'].astype('int64') 
df_clean.info()
```

```python
# access the first five rows of the data frame

df_clean.head()
```

```python
# change data type for one or more columns

df_clean['released'] = df_clean['released'].astype(str)
```

```python
# add a new column derived from an existing column

df_clean['year_correct'] = df_clean['released'].str.extract(pat = '([0-9]{4})').astype('int64')
df_clean['year_correct']
```

```python
# test whether two columns contain the same values

df_clean['year_correct'].equals(df_clean['year'])
```

```python
# remove selected columns

df_clean = df_clean.drop(columns = 'year', axis = 1)
df_clean = df_clean.drop(columns = 'released', axis = 1)
df_clean.head()
```

```python
# get the value of the CPI index

cpi.get(2021)
```

```python
# adjust data for inflation

def inflate_column(data, column):
    return data.apply(lambda x: cpi.inflate(x[column], x.year_correct), axis = 1)

df_clean['adjusted_budget'] = inflate_column(df_clean, 'budget')
df_clean['adjusted_gross'] = inflate_column(df_clean, 'gross')
```

```python
# remove selected columns

df_clean = df_clean.drop(columns = 'budget', axis = 1)
df_clean = df_clean.drop(columns = 'gross', axis = 1)
df_clean.head()
```

```python
# sort a data frame in ascending or descending order

df_clean = df_clean.sort_values(by = ['adjusted_gross'], ascending = False)
df_clean.head()
```

```python
# drop any duplicates

df_clean_duplicates = df_clean.drop_duplicates()
len(df_clean_duplicates)
```

```python
# get the number of rows of a data frame

len(df_clean)
```

```python
# find the unique values within a column and count the number of occurrences of each unique value

unique, counts = np.unique(df_clean['genre'], return_counts=True)
dict(zip(unique, counts))
```

```python
# find the unique values within a column and count the number of occurrences of each unique value

unique, counts = np.unique(df_clean['rating'], return_counts=True)
dict(zip(unique, counts))
```

```python
# specify different replacement values for different existing values

df_clean['rating'].replace(to_replace='NC-17', value= 'X', inplace=True)
df_clean['rating'].replace(to_replace='TV-MA', value= 'X', inplace=True)
df_clean['rating'].replace(to_replace='Unrated', value= 'Not Rated', inplace=True)
df_clean['rating'].replace(to_replace='Approved', value= 'G', inplace=True)
unique, counts = np.unique(df_clean['rating'], return_counts=True)
dict(zip(unique, counts))
```

```python
# plot data and a linear regression model fit

plt.figure(dpi = 100)
sns.regplot(x = 'adjusted_budget', y = 'adjusted_gross', data = df_clean, scatter_kws = {'color': 'blue'}, line_kws = {'color':'red'})
plt.title('Film Budget vs Gross Revenue')

plt.figure(dpi = 100)
sns.regplot(x = 'votes', y = 'adjusted_gross', data = df_clean, scatter_kws = {'color': 'blue'}, line_kws = {'color':'red'})
plt.title('Number of Votes vs Gross Revenue')

plt.figure(dpi = 100)
sns.regplot(x = 'score', y = 'adjusted_gross', data = df_clean, scatter_kws = {'color': 'blue'}, line_kws = {'color':'red'})
plt.title('IMDB Score vs Gross Revenue')

plt.figure(dpi = 100)
sns.regplot(x = 'runtime', y = 'adjusted_gross', data = df_clean, scatter_kws = {'color': 'blue'}, line_kws = {'color':'red'})
plt.title('Length of Film vs Gross Revenue')

plt.figure(dpi = 100)
sns.regplot(x = 'year_correct', y = 'adjusted_gross', data = df_clean, scatter_kws = {'color': 'blue'}, line_kws = {'color':'red'})
plt.title('Year Released vs Gross Revenue')
```

```python
# find the pairwise correlation of all columns in the dataframe

df_clean.corr(method = 'pearson') #pearson, kendall, sperman
```

```python
# plot the heatmap

plt.figure(dpi = 100)

correlation_matrix = df_clean.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
